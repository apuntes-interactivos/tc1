**Ejercicio 1:**  Un sistema trifásico de tres conductores, alimenta una carga en triángulo compuesta por tres impedancias iguales. Determinar las corrientes de fase y de línea, y trazar el diagrama fasorial. Donde $V_{AC}=150\ /240^{\circ}V$ $V_{CB}=150\ /0^{\circ}V$ $V_{BA}=150\ /120^{\circ}V$ $Z_c=25/60^{\circ}\Omega$

```python
#
import lcapy as ly

ce1=ly.Circuit()
ce1.add("""
Z1 A B; rotate=225, size=1.5, l=$Z_c$
W1 A _A1; left, size=1.5
Z2 A C; rotate=315, size=1.5, l=$Z_c$
W2 C _c; down, size=0.4
W3 _c _c2; left, size=2.53
Z3 B C;  l=$Z_c$
W4 B _b; left, size=0.42
""")
ce1.draw()
```
